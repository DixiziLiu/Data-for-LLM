# Check on augment levels

This notebook loads the augments that have been created so far and shows how many augments of each model/temp have been generated.

```python
import pandas as pd
from IPython.display import display, HTML
```

```python
# Read csvs
dfs = [pd.read_csv(f'./data/stratified_data_splits/{str(i+1)}/augments.csv') for i in range(4)]

# Drop empty rows and rows where text is just a bunch of # signs
dfs = [df[df.text.notna() & (~df.text.str.contains("^#+$", na=False, regex=True))] for df in dfs]
```

```python
for i,df in enumerate(dfs):
    # Group by model_id and temperature and count the size of each group
    counts = df.groupby(['model_id', 'temperature']).size().unstack().fillna(0).astype(int)
    
    # Display the result
    display(HTML(f'<h1>Stratified training fold {i+1}'))
    display(counts)
```
